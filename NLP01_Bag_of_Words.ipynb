```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import lxml
```

```python
url = "https://en.wikipedia.org/wiki/Robot"

headers = {
    "User-Agent": "Mozilla/5.0"
}

response = requests.get(url, headers=headers)

soup = BeautifulSoup(response.text, "lxml")



```

```python
content = soup.find_all('p')
with open('robot.txt', "w") as f:
  for i in content:
    f.write(i.text)
```

## Applying Tokenization on text using NLTK

```python
import nltk
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
```

```python
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
```

```python
file = open("robot.txt", "r")
content = file.read()
file.close()
```

```python
sentence = nltk.sent_tokenize(content)
```

```python
import re
corpus = []
for i in range(len(sentence)):
  review = re.sub("[^a-zA-Z]", " ", sentence[i])
  review = review.lower()
  corpus.append(review)
```

```python
corpus
```

## Apply Stemming

```python
for i in corpus:
  words = nltk.word_tokenize(i)
  for word in words:
    if word not in set(stopwords.words("english")):
      print(stemmer.stem(word))
```

## Feature Extraction

```python
processed_corpus = []
for sentence_text in corpus:
    words = nltk.word_tokenize(sentence_text)
    filtered_words = []
    for word in words:
        if word not in stopwords.words('english'):
            filtered_words.append(stemmer.stem(word))
    processed_corpus.append(" ".join(filtered_words))

corpus = processed_corpus
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(binary=True)

x = cv.fit_transform(corpus)
```
